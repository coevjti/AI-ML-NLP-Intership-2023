```python
import spacy
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load the pre-trained word vectors from spaCy
nlp = spacy.load("en_core_web_md")

# Prepare your word data
# words = ["apple", "banana", "orange", "car", "bike", "bus"]
words=s

# Convert words into their corresponding vectors
word_vectors = [nlp(word).vector for word in words]

# Choose the number of clusters for K-means
num_clusters = 20

# Apply K-means clustering
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(word_vectors)

# Get cluster labels and print the clusters
cluster_labels = kmeans.labels_
clusters = {}
for i, word in enumerate(words):
    cluster = cluster_labels[i]
    if cluster not in clusters:
        clusters[cluster] = []
    clusters[cluster].append(word)

for cluster, words in clusters.items():
    print("Cluster:", cluster)
    print("Words:", words)
    print()

# Apply dimensionality reduction for visualization
pca = PCA(n_components=2)
reduced_embeddings = pca.fit_transform(word_vectors)

# Plot the word clusters
plt.scatter(reduced_embeddings[:, 0], reduced_embeddings[:, 1], c=cluster_labels)
for i, word in enumerate(words):
    plt.annotate(word, (reduced_embeddings[i, 0], reduced_embeddings[i, 1]))
plt.show()
```

# **Optimizing Methods of K-mean Clustering**

```python
import spacy
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the pre-trained word vectors from spaCy
nlp = spacy.load("en_core_web_md")

words = s

# Convert words into their corresponding vectors
word_vectors = [nlp(word).vector for word in words]

# Define the range of K values to try
k_values = range(1, 100)

# Calculate the sum of squared distances for each K
sse = []
for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(word_vectors)
    sse.append(kmeans.inertia_)

# Plot the sum of squared distances vs. K
plt.plot(k_values, sse)
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Sum of Squared Distances")
plt.title("Elbow Method")
plt.show()
```

```python
import spacy
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Load the pre-trained word vectors from spaCy
nlp = spacy.load("en_core_web_md")

words = s

# Convert words into their corresponding vectors
word_vectors = [nlp(word).vector for word in words]

# Define the range of K values to try
k_values = range(3, 50)

# Calculate silhouette score for each K
silhouette_scores = []
for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(word_vectors)
    labels = kmeans.labels_
    silhouette_scores.append(silhouette_score(word_vectors, labels))

# Find the optimal K with the highest silhouette score
optimal_k = k_values[silhouette_scores.index(max(silhouette_scores))]
print("Number of Clusters to be formed : ", optimal_k)
print()

# Apply K-means clustering with the optimal K
kmeans = KMeans(n_clusters=optimal_k)
kmeans.fit(word_vectors)

# Get cluster labels and print the clusters
cluster_labels = kmeans.labels_
clusters = {}
for i, word in enumerate(words):
    cluster = cluster_labels[i]
    if cluster not in clusters:
        clusters[cluster] = []
    clusters[cluster].append(word)

for cluster, words in clusters.items():
    print("Cluster:", cluster)
    print("Words:", words)
    print()
```

```python
from sentence_transformers import SentenceTransformer
import hdbscan
import umap
import matplotlib.pyplot as plt

# Load the pre-trained Sentence Transformer model
model = SentenceTransformer('bert-base-nli-mean-tokens')

words = s
# Convert words into their corresponding sentence embeddings
sentence_embeddings = model.encode(words)

# Apply UMAP for dimensionality reduction
reducer = umap.UMAP(n_neighbors=4, n_components=3)
umap_embeddings = reducer.fit_transform(sentence_embeddings)

# Apply HDBSCAN clustering
min_cluster_size = 5
min_samples = 3
clusterer = hdbscan.HDBSCAN(min_cluster_size=min_cluster_size, min_samples=min_samples)
cluster_labels = clusterer.fit_predict(umap_embeddings)

# Get cluster labels and print the clusters (excluding outliers)
clusters = {}
for i, word in enumerate(words):
    cluster = cluster_labels[i]
    if cluster != -1:  # Exclude outliers
        if cluster not in clusters:
            clusters[cluster] = []
        clusters[cluster].append(word)

# Plot the UMAP visualization with cluster colors
plt.scatter(umap_embeddings[:, 0], umap_embeddings[:, 1], c=cluster_labels, cmap='viridis', s=5)
plt.colorbar()
plt.show()

for cluster, words in clusters.items():
    print("Cluster:", cluster)
    print("Words:", words)
    print()
```
