```python
import requests
from bs4 import BeautifulSoup

# Send a GET request to the website
url = "https://attack.mitre.org/tactics/TA0043/"
response = requests.get(url)

# Parse the HTML content using Beautiful Soup
soup = BeautifulSoup(response.text, "html.parser")

# Find all <a> tags that contain links to PDF files
pdf_links = soup.find_all("a", href=lambda href: href and href.endswith(".pdf"))

# Print the PDF links
for link in pdf_links:
    pdf_url = link.get("href")
    print(pdf_url)
```

```python
import requests
import pdfplumber
import json
import tempfile
from pprint import pprint

# Send a GET request to the PDF file
pdf_url = "http://www.esahity.com/uploads/5/0/1/2/501218/malvani_rang_arun_kulkarni.pdf"
response = requests.get(pdf_url)

# Save the PDF content as a temporary file
with tempfile.NamedTemporaryFile(suffix=".pdf", delete=False) as temp_file:
    temp_file.write(response.content)
    temp_file.close()

    # Load the PDF content using pdfplumber
    with pdfplumber.open(temp_file.name) as pdf:
        extracted_text = ""
        for page in pdf.pages:
            extracted_text += page.extract_text()

# Create a JSON object with the extracted text
json_data = {
    "text": extracted_text
}

# # Convert the JSON object to a string
# json_string = json.dumps(json_data, ensure_ascii=False)

# # Print the JSON string
# print(json_string, end="")

# Pretty-print the JSON object
pprint(json_data, width=100)

```
